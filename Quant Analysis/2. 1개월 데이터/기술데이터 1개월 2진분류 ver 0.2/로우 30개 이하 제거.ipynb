```python
non_out_kospi_200_items = pd.read_excel('./non_out_data.xlsx')
```

```python
def data_sp(data):
    com_name_set = list(set(data.Name))
    
    ## 18년도 3,6월 이전 이후로 나누기
    train_set = pd.DataFrame()
    test_set = pd.DataFrame()
    for i in range(len(com_name_set)):
        sub_set = data[ data.Name == com_name_set[i]]
        train_set = train_set.append(sub_set[:-6])
        test_set = test_set.append(sub_set[-6:])
    return train_set , test_set
```

```python
ls = []
for i in list(set(non_out_kospi_200_items.Name)):
    if len(non_out_kospi_200_items[non_out_kospi_200_items.Name == i]) < 50:
        ls.append(i)
```

```python
## train test를 X,Y 나누기
X_train = train[col]

X_test = test[col]

Y_train = train['1개월 수익률(2)']
Y_test = test['1개월 수익률(2)']
```

```python
from numpy import loadtxt
from xgboost import  XGBClassifier
from xgboost import plot_importance
import xgboost
from sklearn import metrics

model = XGBClassifier()
model.fit(X_train ,Y_train)
plt.figure(figsize=(40,30))
plot_importance(model)
# plt.show()
a = model.predict(X_test)
print(metrics.accuracy_score(a,Y_test))
t = metrics.confusion_matrix(Y_test,a)
print(t,t[1,1]/t[:,1].sum())
```

```python
acc = []
pre = []
for i in [2,3,4,5,6,7]:
    for k in [0,2,4,6,8,10,15,20]:
        for j in [0,1,2,4,6,8,10,20,30]:
            model = XGBClassifier(max_depth=i, gamma=k,min_child_weight = j)
            model.fit(X_train ,Y_train)
            a = model.predict(X_test)
            t = metrics.confusion_matrix(Y_test,a)
            print('max_depth:{} , gamma: {},  min_child_weight:{} '.format(i,k,j),'\n',t,'\n','pre',t[1,1]/t[:,1].sum(),'\n','acc',
                  metrics.accuracy_score(a,Y_test),'\n','**************************************')
            acc.append(metrics.accuracy_score(a,Y_test))
            pre.append(t[1,1]/t[:,1].sum())
```
