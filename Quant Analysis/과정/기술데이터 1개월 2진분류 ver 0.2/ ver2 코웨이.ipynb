```python
def presnet_to_past(data,col):
    sets = pd.DataFrame()
    for i in list(set(data.Name)):
        sub = data[data.Name==i][col]
        present0 = sub[1:]
        pre0 = sub[:-1]
        present0.index = pre0.index
        c = present0 > pre0
        aa = pd.concat([pd.DataFrame([0]), c])
        aa['Name']= i
        aa.index = sub.index
        sets = sets.append(aa[1:])
    sets.columns = [col,'Name']
    return sets
```

```python
def cate_2(data,col):
    sets = pd.DataFrame()
    for i in list(set(data.Name)):
        sub = data[data.Name==i][col]
        aa = pd.DataFrame(sub >0)
        aa['Name']= i
        sets = sets.append(aa[1:])
        sets[col] = sets[col].astype(np.int)
    return sets
```

```python
def data_sp(data):
    com_name_set = list(set(data.Name))
    
    train_set = pd.DataFrame()
    test_set = pd.DataFrame()
    for i in range(len(com_name_set)):
        sub_set = data[ data.Name == com_name_set[i]]
        train_set = train_set.append(sub_set[:-30])
        test_set = test_set.append(sub_set[-30:])
    return train_set , test_set
```
