```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
home = 'https://finance.naver.com'
url = home+'/sise/sise_group.nhn?type=upjong'
pageInfo = urlopen( url )
soup = BeautifulSoup( pageInfo,  'html.parser')
```

```python
print( soup.prettify() )
```

```python
##https://finance.naver.com/sise/sise_group.nhn?type=upjong
sets = soup.findAll(['td','a'],style="padding-left:10px;")
```

```python
#업종 이름 태그는 .태그네임
sets[0].a.string
## 링크 / =으로 끝나면 ['stm']
sets[0].a['href']
```

```python
sets
```

```python
sectorname = []
addr = []

for set in sets:
    sectorname.append(set.a.string)
    addr.append(set.a['href'])
```

```python
for i in range(len(addr)):
    addr[i] = home+addr[i]
```

```python
sectordict = dict(zip(sectorname,addr))
```

```python
pageInfo = urlopen( sectordict[sectorname[0]]) 
soup1 = BeautifulSoup( pageInfo,  'html.parser')
```

```python
sets1 = soup1.findAll('table')
```
